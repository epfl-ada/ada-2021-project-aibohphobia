# Analyse the categories

```python
def create_df_categories_proportion(df):
    df_tags_explode = df.explode("tags").explode("tags")
    df_tags = df_tags_explode.groupby(["tags","gender"]).quoteID.count().to_frame(name="Count").sort_values(['Count'],ascending=False).reset_index()
    df_total_per_cat = df_tags.groupby(['tags']).Count.sum().to_frame(name="Total").reset_index()
    df_tags = df_tags.merge(df_total_per_cat, on='tags')
    df_tags['Proportion']  = (df_tags['Count']/df_tags['Total'])*100
    return df_tags
```

```python
def plot_quotes_categories_proportion(df_tags):
    """
        Function to plot the number of quotes depending on `tags` and `gender`
    :param df: dataframe 
    """
    #create plot
    f = plt.figure(figsize=(18,6))
    ax = sns.barplot(data=df_tags, x='tags',y='Proportion', hue='gender')
    plt.xlabel('Category')
    plt.ylabel('Proportion of quotes')
    plt.title('Proporion of quotes between genders, per category')
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=45,  horizontalalignment='right')
```

# Analyse categories over years

```python
# Open

with open('df_2015_categories.pkl', 'rb') as file:
    df_2015_cat = pickle.load(file)

with open('df_2016_categories.pkl', 'rb') as file:
    df_2016_cat = pickle.load(file)
    
with open('df_2017_categories.pkl', 'rb') as file:
    df_2017_cat = pickle.load(file)
    
with open('df_2018_categories.pkl', 'rb') as file:
    df_2018_cat = pickle.load(file)
    
with open('df_2019_categories.pkl', 'rb') as file:
    df_2019_cat = pickle.load(file)
    
with open('df_2020_categories.pkl', 'rb') as file:
    df_2020_cat = pickle.load(file)
```

```python
# Concatenate all tables 

df_2015_cat['year'] = [2015]*len(df_2015_cat)
df_2016_cat['year'] = [2016]*len(df_2016_cat)
df_2017_cat['year'] = [2017]*len(df_2017_cat)
df_2018_cat['year'] = [2018]*len(df_2018_cat)
df_2019_cat['year'] = [2019]*len(df_2019_cat)
df_2020_cat['year'] = [2020]*len(df_2020_cat)

frames = [df_2015_cat, df_2016_cat, df_2017_cat, df_2018_cat, df_2019_cat, df_2020_cat]
df_cat = pd.concat(frames)
```

```python
df_cat
```

```python
# Plot category evolution over years

for cat in df_cat['tags'].unique():
    rows_cat = df_cat[df_cat['tags']==cat]
    
    f = plt.figure(figsize=(10,3))
    ax = sns.barplot(data=rows_cat, x='year',y='Proportion', hue='gender')
    plt.ylabel('Proportion of quotes')
    plt.title('Proporion of quotes between genders in %s' %cat)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=45,  horizontalalignment='right')
    
```

# Analysing gap between genders

```python
df_cat['diff_prop'] = df_cat.Proportion.diff()
```

```python
df_cat.reset_index(drop=True, inplace=True)
```

```python
# We just want the difference male - female per categpry and year
df_cat.loc[df_cat["gender"]=='Male', "diff_prop"] = np.nan
df_cat.loc[df_cat["gender"]=='Female', "diff_prop"] *= -1
```

```python
df_cat
```

```python
for cat in df_cat['tags'].unique():
    rows_cat = df_cat[df_cat['tags']==cat]
    rows_plot = rows_cat[rows_cat['gender']=='Female']
    
    f = plt.figure(figsize=(10,3))
    ax = sns.barplot(data=rows_plot, x='year',y='diff_prop')
    plt.ylabel('Difference in proportion of quotes')
    plt.title('Difference in quote proportion in %s' %cat)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=45,  horizontalalignment='right')
```

# Evolution in time

```python
# difference in each category, over years

for cat in df_cat['tags'].unique():
    rows_cat = df_cat[df_cat['tags']==cat]
    rows_plot = rows_cat[rows_cat['gender']=='Female']
    
    f = plt.figure(figsize=(10,3))
    sns.lineplot(data=rows_plot, x='year',y='diff_prop')
    plt.ylim(0,100)
    plt.ylabel('Difference in proportion')
    plt.title('Difference in quote proportion in %s' %cat)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=45,  horizontalalignment='right')
```

```python
# difference in each category, over years per gender

for cat in df_cat['tags'].unique():
    rows_plot = df_cat[df_cat['tags']==cat]
        
    f = plt.figure(figsize=(10,3))
    sns.lineplot(data=rows_plot, x='year',y='Proportion', hue='gender')
    plt.ylim(0,100)
    plt.ylabel('Quote proportion between genders over time')
    plt.title('Quote proportion in %s' %cat)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=45,  horizontalalignment='right')
```
